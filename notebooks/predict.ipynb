```python
# Configuration for Prediction
INPUT_IMAGE_PATH = "path/to/your/image.png" # REPLACE WITH YOUR IMAGE PATH
OPACITY = 0.9

# Optional: Specify a trained checkpoint 
# (if you not specify it, the default checkpoint in the config file will be used)
CHECKPOINT_FILE = "best.pth" # REPLACE WITH YOUR CHECKPOINT NAME
CHECKPOINT_PATH = os.path.join(BASE_PATH, "finestSAM", "sav", CHECKPOINT_FILE)
MODEL_TYPE = "vit_b" # REPLACE WITH YOUR MODEL TYPE
```
